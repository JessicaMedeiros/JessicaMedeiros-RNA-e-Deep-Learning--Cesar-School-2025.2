# Regressão Linear com PyTorch

## Objetivo

Este exemplo utiliza regressão linear para estimar o comprimento das sépalas da íris a partir do seu comprimento das pétalas.
Utiliza-se
- o cálculo do gradiente automático do PyTorch, porém
- a função de perda e
- atualização dos parâmetros é feita de forma explícita.
- A rede é criada com uma camada nn.Linear()

![alt text](https://raw.githubusercontent.com/vcasadei/images/master/RegressaoLinear_Otimizacao.png)

## Importação dos pacotes

```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
from sklearn.datasets import load_iris

import torch
from torch import nn, optim
from torch.autograd import Variable

torch.manual_seed(1234)
```

## Leitura dos dados

```python
iris = load_iris()
data = iris.data[iris.target==1,::2]  # comprimento das sépalas e pétalas, indices 0 e 2

x_train = data[:,0:1].astype(np.float32)
y_train = data[:,1:2].astype(np.float32)

n_samples = x_train.shape[0]
print('x_train.shape:',x_train.shape, x_train.dtype)
print('y_train.shape:',y_train.shape, y_train.dtype)

print('x_train[:5]:\n', x_train[:5])
print('y_train[:5]:\n', y_train[:5])
```

### Normalização dos dados

```python
x_train -= x_train.min()
x_train /= x_train.max()
y_train -= y_train.min()
y_train /= y_train.max()
```

```python
print('x_train[:5]:\n', x_train[:5])
print('y_train[:5]:\n', y_train[:5])
```

```python
x_train_bias = np.hstack([np.ones(shape=(n_samples,1)), x_train])
```

```python
x_train_bias
```

```python
print('x_train_bias.shape:',x_train_bias.shape, x_train_bias.dtype)
```

```python
x_train_bias = torch.FloatTensor(x_train_bias)
y_train      = torch.FloatTensor(y_train)
print(y_train)
```

## Model

![alt text](https://raw.githubusercontent.com/vcasadei/images/master/RegressaoLinearNeuronio.png)

```python
#Inicialização usando toch.nn.Linear

model = torch.nn.Linear(in_features=2, out_features=1, bias=False)
```

### Verificando a inicialização dos parâmetros

```python
model.weight.data = torch.zeros(1,2)
torch.nn.init.uniform(model.weight.data, -0.1, 0.1)
model.weight
```

### Testando o predict da rede

```python
xx = torch.ones((5,2))
xx
```

```python
model(Variable(xx))
```

## Treinamento

```python
num_epochs = 100
learning_rate = 0.5
for epoch in range(num_epochs):
    #inputs = Variable(x_train_bias)
    #target = Variable(y_train)
    inputs = x_train_bias
    target = y_train

    # forward - predict
    y_pred = model(inputs)

    #loss cálculo da função de perda
    loss = torch.mean((y_pred - target)**2)

    # backward propagation - cálculo gradient
    loss.backward()

    # gradiente descendente
    model.weight.data = model.weight.data - learning_rate * model.weight.grad.data

    # zeragem parâmetros próximo cálculo do gradiente
    model.weight.grad.data.zero_()

    # verbose
    if (epoch+1) % 20 == 0:
        print('Epoch[{}/{}], loss: {:.6f}'
              .format(epoch+1, num_epochs, loss.data))
```

## Avaliação

```python
y_pred = model(Variable(x_train_bias))
plt.plot(x_train, y_train.numpy(), 'ro', label='Original data')
plt.plot(x_train, y_pred.data.numpy(), 'kx-', label='Fitting Line')
plt.show()
```

### Acessando os parâmetros

```python
w_dic = model.state_dict()
w_dic
```

### Parâmetros ótimos, solução analítica

```python
x_bias = x_train_bias
y = y_train
x_bias_t = torch.t(x_bias)
w_opt = (torch.inverse(x_bias_t.mm(x_bias)).mm(x_bias_t)).mm(y)
print(torch.t(w_opt))
```

# Aprendizados


1 - Por que no final da varredura do laço de treinamento, a perda é impressa utilizando loss.data? Por que precisa do data?

A variável loss no PyTorch é um Tensor preso a um histórico de operações (grafo computacional). Usamos o .data para extrair apenas o valor numérico puro. Isso é essencial para:

Evitar sobrecarga de memória (RAM): Impede que o PyTorch acumule o histórico de cálculos desnecessariamente a cada época.

Limpar o print: Exibe apenas o número do erro, sem informações extras do tensor.

Atualização Importante: O .data ficou obsoleto. Nas versões modernas do PyTorch, a forma correta de extrair esse valor é utilizando .item() (ex: loss.item()).

```python
# 2 - Calcule e imprima o valor da perda (MSE) fazendo o predict da rede com os pesos já treinados.

y_pred_final = model(x_train_bias)

loss_final = torch.mean((y_pred_final - y_train) ** 2)

print(f"Valor da perda (MSE) final: {loss_final.item():.6f}")
```

```python
# Inicialização das listas
w0_list = []
w1_list = []

model.weight.data = torch.zeros(1, 2)
torch.nn.init.uniform_(model.weight.data, -0.1, 0.1)

num_epochs = 100
learning_rate = 0.5

for epoch in range(num_epochs):
    inputs = x_train_bias
    target = y_train

    w0_list.append(model.weight.data[0][0].item())
    w1_list.append(model.weight.data[0][1].item())

    # forward - predict
    y_pred = model(inputs)

    # loss cálculo da função de perda
    loss = torch.mean((y_pred - target)**2)

    loss.backward()

    model.weight.data -= learning_rate * model.weight.grad.data

    model.weight.grad.data.zero_()


x_bias_t = torch.t(x_train_bias)
w_opt = (torch.inverse(x_bias_t.mm(x_train_bias)).mm(x_bias_t)).mm(y_train)


w0_otimo = w_opt[0][0].item()
w1_otimo = w_opt[1][0].item()

plt.figure(figsize=(10, 7))

plt.scatter(w0_list, w1_list, c='blue', alpha=0.5, label='Evolução dos Pesos')

w0_old = None
for (w0, w1) in zip(w0_list, w1_list):
    if w0_old is not None:
        plt.arrow(w0_old, w1_old, w0 - w0_old, w1 - w1_old,
                  head_length=0.015, head_width=0.015, shape='full',
                  length_includes_head=True, color='gray', alpha=0.4)
    w0_old, w1_old = w0, w1


plt.plot(w0_otimo, w1_otimo, 'r*', label='Solução Analítica (Ótima)')

plt.xlabel('w0 (Peso da coluna de 1s / Bias)')
plt.ylabel('w1 (Peso do Comprimento das Pétalas)')
plt.title('Evolução dos Parâmetros do Modelo (w0 vs w1)')
plt.legend()
plt.grid(True)
plt.show()
```
